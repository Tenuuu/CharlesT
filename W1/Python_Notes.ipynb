Python is a dynamically typed language

```python
x = 10 # numerical
print(x)
x = "hello there" # string
print(x)
x = True # boolean
print(x)
x = [1,2,3] # list
print(x)
x = None 
print(x)
```

## Python Operators
### Arithmetic
+ - addition
- : subtraction
* - multiplicaton
/ - division
% - remainder/modulus
// - integer division / floor division
** - exponentiation

### Logical
> - greater than
< - less than
== - logical equal to
>= - greater or equal to
<= - less than or equal to
!= - NOT equal to
and - logically both statement must be True to produce a True result
or - logially either or both statements can be True to produce a True result
not - logical inversion - makes True False and makes False True

### Assignment Ops
= - simple assignment operator
+= - shorthand addition and assignment
-= - shorthand subtraction and assignment
*= - shorthand multiply and assignment
/= - shorthand divide and assignment

### Membership
'in' - True if a value is a member of a collection
'not in' - True if a value is not a member of a collection

### Identity ops
'is' - True if two variables refer to the same object in memory
'is not' - True if two variables do not refer to the same object in memory

### Order of Precedence
- arithmatic ops
- comparisons
- logical (and, or, not)

use parentheses to control the execution more carefully.

## Control Flow

### If-Else
decides to exececute a block of code based on a condition

```python
if True == False:
    print("Yep, that's True!")
elif False == False:
    print("Yep, that's True too!!")
else:
    print("Nope, that's False!")
```

### Loops
for - iterate a set number of times - tracking what iteration you're on!
while - conditional - less about the iteration - keep going until the condition is false
do-while - conditional, but runs the code at least once before checking the condition

```python
for x in range(6, 12, 2):
    print(x)
```

```python

userInput = "a"
while True:

    if userInput == "q":
        break
    print("Hello World")
    userInput = input()
    
```

Always try to write DRY code: Don't Repeat Yourself

```python
def myFunction(value):
    print(value)

def addition(a, b):
    return a + b

myFunction(5)
myFunction("Hello")
myFunction(True)

x = addition(2, 3)
print(x)
```
